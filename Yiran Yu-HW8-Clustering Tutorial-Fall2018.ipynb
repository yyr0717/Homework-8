1. A short description of the algorithm: 

DBSCAN (Density-based Noise Application Spatial Clustering) is a popular clustering algorithm, which is used to replace K-means in prediction analysis. It does not require me to enter the number of clusters to run. But in exchange, I have to adjust the other two parameters.

2. Data setup：  

We use the Iris dataset for the first prediction. The data set contains a set of 150 records, belonging to five attributes - petal length, petal width, sepal length, sepal width and category. Iris Setosa, Iris Virginica and Iris Versicolor are three classes. We use sklearn Library in Python to load Iris dataset, and matplotlib for data visualisation.

```python
# Importing Modules
from sklearn import datasets
import matplotlib.pyplot as plt

# Loading dataset
iris_df = datasets.load_iris()

# Available methods on dataset
print(dir(iris_df))

# Features
print(iris_df.feature_names)

# Targets
print(iris_df.target)

# Target Names
print(iris_df.target_names)
label = {0: 'red', 1: 'blue', 2: 'green'}

# Dataset Slicing
x_axis = iris_df.data[:, 0]  # Sepal Length
y_axis = iris_df.data[:, 2]  # Sepal Width

# Plotting
plt.scatter(x_axis, y_axis, c=iris_df.target)
plt.show()
```

3. Algorithm run ：
4. Display in graph and plots ：

```python
# Importing Modules
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA

# Load Dataset
iris = load_iris()

# Declaring Model
dbscan = DBSCAN()

# Fitting
dbscan.fit(iris.data)

# Transoring Using PCA
pca = PCA(n_components=2).fit(iris.data)
pca_2d = pca.transform(iris.data)

# Plot based on Class
for i in range(0, pca_2d.shape[0]):
    if dbscan.labels_[i] == 0:
        c1 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='r', marker='+')
    elif dbscan.labels_[i] == 1:
        c2 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='g', marker='o')
    elif dbscan.labels_[i] == -1:
        c3 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='b', marker='*')

plt.legend([c1, c2, c3], ['Cluster 1', 'Cluster 2', 'Noise'])
plt.title('DBSCAN finds 2 clusters and Noise')
plt.show()
```

5. Discussion of the images in the graphs and plots: 

The scikit-learn implementation provides default values for EPS and min_samples parameters, but I am usually expected to adjust these parameters. The EPS parameter is the maximum distance between the two data points to be considered in the same neighborhood. The min_samples parameter is the minimum number of data points that are considered as clusters in the neighborhood.
